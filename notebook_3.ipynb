````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—á–∏—Å—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö
def clean_data(df):
    # –£–¥–∞–ª—è–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã –≤ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].str.strip()
    
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–µ–∫—Å—Ç–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].str.lower()
    
    return df

# –û—á–∏—â–∞–µ–º –∫–∞–∂–¥—ã–π DataFrame –≤ —Å–ø–∏—Å–∫–µ
for i, data in enumerate(tables_list):
    tables_list[i] = clean_data(data)

# –ü–æ–∏—Å–∫ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
print("–ü–æ–∏—Å–∫ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤:")
for data in tables_list:
    print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {data.duplicated().sum()}")

# –ü–æ–∏—Å–∫ —Å–∫—Ä—ã—Ç—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º —Å—Ç–æ–ª–±—Ü–∞–º
print("\n–ü–æ–∏—Å–∫ —Å–∫—Ä—ã—Ç—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º —Å—Ç–æ–ª–±—Ü–∞–º:")
for data in tables_list:
    categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã
    if len(categorical_columns) > 0:
        print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–∫—Ä—ã—Ç—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {data.duplicated(subset=categorical_columns).sum()}")
    else:
        print("–ù–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤.")

# –ü–æ–∏—Å–∫ –ø—Ä–æ–ø—É—Å–∫–æ–≤
print("\n–ü–æ–∏—Å–∫ –ø—Ä–æ–ø—É—Å–∫–æ–≤:")
for data in tables_list:
    print(data.isna().sum())

# –û–±–∑–æ—Ä —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
print("\n–û–±–∑–æ—Ä —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:")
for data in tables_list:
    print(data.nunique())
````

## 3. –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

–®–∞–≥ 1: –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∞–∫—Ç–∏–≤–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ (‚â•3 –ø–µ—Ä–∏–æ–¥–∞ —Å –ø–æ–∫—É–ø–∫–∞–º–∏)
–ò—Å–ø–æ–ª—å–∑—É–µ–º market_money.csv, –≥–¥–µ –µ—Å—Ç—å –¥–∞–Ω–Ω—ã–µ –æ –≤—ã—Ä—É—á–∫–µ –ø–æ –ø–µ—Ä–∏–æ–¥–∞–º.

````python

# –§–∏–ª—å—Ç—Ä—É–µ–º –∑–∞–ø–∏—Å–∏ —Å –Ω–µ–Ω—É–ª–µ–≤–æ–π –≤—ã—Ä—É—á–∫–æ–π –∏ —Å—á–∏—Ç–∞–µ–º —á–∏—Å–ª–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–µ—Ä–∏–æ–¥–æ–≤ –Ω–∞ –∫–ª–∏–µ–Ω—Ç–∞
active_clients = (
    market_money[market_money["–≤—ã—Ä—É—á–∫–∞"] > 0]
    .groupby("id")["–ø–µ—Ä–∏–æ–¥"]
    .nunique()
    .loc[lambda x: x >= 3]
    .index
)
print(f"–ö–ª–∏–µ–Ω—Ç–æ–≤ —Å –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å—é ‚â•3 –º–µ—Å—è—Ü–µ–≤: {len(active_clients)}")
print(type(active_clients))
````

–®–∞–≥ 2: –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü –ø–æ –∞–∫—Ç–∏–≤–Ω—ã–º –∫–ª–∏–µ–Ω—Ç–∞–º

````python

# –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è
market_file = market_file[market_file["id"].isin(active_clients)]
market_money = market_money[market_money["id"].isin(active_clients)]
market_time = market_time[market_time["id"].isin(active_clients)]
money = money[money["id"].isin(active_clients)]
# –°–ø–∏—Å–æ–∫ —Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
filtered_data_list = [market_file, market_money, market_time, money]
# –í—ã–≤–æ–¥ –¥–ª–∏–Ω—ã –∫–∞–∂–¥–æ–≥–æ —Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞ –¥–ª—è –æ—Ä–∏–µ–Ω—Ç–∏—Ä–∞
print([len(i) for i in filtered_data_list], sep="\n")
print(*[i.head(3) for i in filtered_data_list])
````

````python


market_file_num_columns = market_file.drop(columns=['id'], errors='ignore') \
                                   .select_dtypes(include=['int64', 'float64']).columns.tolist()
discrette_columns = ['–º–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_—Ç–µ–∫_–º–µ—Å', '—Å—Ä–µ–¥–Ω–∏–π_–ø—Ä–æ—Å–º–æ—Ç—Ä_–∫–∞—Ç–µ–≥–æ—Ä–∏–π_–∑–∞_–≤–∏–∑–∏—Ç', '–Ω–µ–æ–ø–ª–∞—á–µ–Ω–Ω—ã–µ_–ø—Ä–æ–¥—É–∫—Ç—ã_—à—Ç—É–∫_–∫–≤–∞—Ä—Ç–∞–ª', '–æ—à–∏–±–∫–∞_—Å–µ—Ä–≤–∏—Å–∞']
market_money_num_columns = market_money.drop(columns='id', errors='ignore') \
                                   .select_dtypes(include=['int64', 'float64']).columns.tolist()
market_time_num_columns = market_time.drop(columns='id', errors='ignore') \
                                   .select_dtypes(include=['int64', 'float64']).columns.tolist()
market_file_cat_columns = market_file.drop(columns='id', errors='ignore') \
                                   .select_dtypes('object').columns.tolist()
market_money_cat_columns = market_money.drop(columns='id', errors='ignore') \
                                   .select_dtypes('object').columns.tolist()
market_time_cat_columns = market_time.drop(columns='id', errors='ignore') \
                                   .select_dtypes('object').columns.tolist()
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞–∑–º–∞—Ö–∞ –¥–ª—è —Å—Ç–æ–ª–±—Ü–æ–≤ —Ç–∞–±–ª–∏—Ü
def histogram_boxplot(table_file, label):
    label_common = label.replace('_', ' ').capitalize()
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
    sns.histplot(data=table_file, x=label, ax=axs[0], bins = 40, kde=True)
    axs[0].set_xlabel(label_common)
    axs[0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    axs[0].set_title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞')
    if not label in  discrette_columns:
        sns.boxplot(data=table_file, x=label, ax=axs[1])
        axs[1].set_xlabel(label_common)
        axs[1].set_title('–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞–∑–º–∞—Ö–∞')
    else:
        sns.countplot(data=table_file, x=label, ax=axs[1])
        axs[1].set_xlabel(label_common)
        axs[1].set_title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    fig.suptitle(label_common)
    plt.show()

def pie_plot(table_file, label):
    # label_common = label.replace('_', ' ').capitalize()
    # fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(12, 4))
    # table_file[label].value_counts().plot(y = 'count', kind = 'pie', autopct='%1.0f%%', title = label_common, ax = axs)
    # plt.show()
    palette_color = sns.color_palette('bright') 
    table_file[label].value_counts().plot(y='count', kind='pie', autopct='%1.0f%%', title=label.capitalize(), colors=palette_color, figsize=(14, 6))
    plt.show()


print("–°–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –±–µ–∑ 'id' –≤ market_file:")
print(market_file_num_columns)
print("–°–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –±–µ–∑ 'id' –≤ market_money:")
print(market_money_num_columns)
print("–°–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –±–µ–∑ 'id' –≤ market_time:")
print(market_time_num_columns)
print("–°–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –±–µ–∑ 'id' –≤ market_file:")
print(market_file_cat_columns)
print("–°–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –±–µ–∑ 'id' –≤ market_money:")
print(market_money_cat_columns)
print("–°–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –±–µ–∑ 'id' –≤ market_time:")
print(market_time_cat_columns)
````

## 6. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ø–∞–π–ø–ª–∞–π–Ω–æ–≤

#### 6.1 –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö


````python

# –î–µ–ª–∏–º –Ω–∞ —á–∏—Å–ª–æ–≤—ã–µ –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
print("–ß–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏:", numerical_cols)  # –æ—Ç–æ–±—Ä–∞–Ω—ã –Ω–∞ –ø—Ä–µ–¥—ã–¥—É—â–µ–º —ç—Ç–∞–ø–µ

# –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
categorical_cols = merged_data.select_dtypes(include=['object']).columns.tolist()
categorical_cols = [col for col in categorical_cols if col != 'id' and col != '–ü–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å']
print(categorical_cols)
````

#### 6.2 –î–æ–±–∞–≤–ª—è–µ–º ohe –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö - —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω

````python
ohe_pipe = Pipeline(
    [
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore')
        )
    ]
) 
ord_pipe = Pipeline(
    [
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['–Ω–µ—Ç', '–¥–∞'],
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
    ]
)
````

–û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ ColumnTransformer

````python

# –†–∞–∑–¥–µ–ª—è–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ 2 –≥—Ä—É–ø–ø—ã - –¥–ª—è —Ä–∞–∑–Ω—ã—Ö —Å–∫–µ–π–ª–µ—Ä–æ–≤
ohe_features = ['—Ç–∏–ø_—Å–µ—Ä–≤–∏—Å–∞', '–ø–æ–ø—É–ª—è—Ä–Ω–∞—è_–∫–∞—Ç–µ–≥–æ—Ä–∏—è']  # –í—Å–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –¥–ª—è OneHotEncoder
ord_features = ['—Ä–∞–∑—Ä–µ—à–∏—Ç—å_—Å–æ–æ–±—â–∞—Ç—å']  # –û–¥–Ω–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏—è –¥–ª—è OrdinalEncoder

# –°–æ–∑–¥–∞–µ–º ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', ohe_pipe, ohe_features),  # –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è OneHotEncoder
        ('ord', ord_pipe, ord_features)   # –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è OrdinalEncoder
    ],
    remainder='drop'
)
# –†–∞–∑–¥–µ–ª—è–µ–º –∫–∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ 2 –≥—Ä—É–ø–ø—ã - –¥–ª—è —Ä–∞–∑–Ω—ã—Ö —Å–∫–µ–π–ª–µ—Ä–æ–≤
num_features_group1 = ['–≤—ã—Ä—É—á–∫–∞_–ø—Ä–µ–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü', '–≤—ã—Ä—É—á–∫–∞_–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü', '–¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å']
num_features_group2 = ['–∞–∫—Ü–∏–æ–Ω–Ω—ã–µ_–ø–æ–∫—É–ø–∫–∏', '–º–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_6_–º–µ—Å', '–º–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_—Ç–µ–∫_–º–µ—Å']

# –°–æ–∑–¥–∞–µ–º ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', ohe_pipe, ohe_features),
        ('num_group2', StandardScaler(), num_features_group2),
        ('num_group1', MinMaxScaler(), num_features_group1)
    ],
    remainder='drop'
)
````

–£—Å–ø–µ—Ö üëç:


 
- 'passthrough' –Ω–µ –∑–∞–±—ã—Ç
- drop='first' —É–∫–∞–∑–∞–Ω
- handle_unknown='ignore'/handle_unknown='use_encoded_value'  
- unknown_value=np.nan    
  
–í—ã–≥–ª—è–¥–∏—Ç —Ç–∞–∫ —á—Ç–æ  –æ–±–µ—Å–ø–µ—á–µ–Ω–∞ –±–µ—Å–ø–µ—Ä–µ–±–æ–π–Ω–∞—è —Ä–∞–±–æ—Ç–∞ –Ω–∞—à–µ–≥–æ pipeline    
     

````python
merged_data['–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è_–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'].unique()
merged_data['–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è_–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'].value_counts()
print(merged_data['–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è_–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'].isna().sum())
````

–ò—Ç–æ–≥–æ–≤—ã–π pipeline

````python
# –°–æ–∑–¥–∞–µ–º –ø–æ–ª–Ω—ã–π –ø–∞–π–ø–ª–∞–π–Ω
pipe_final = Pipeline(
    [
        ('preprocessor', preprocessor),
        ('model', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é
X = merged_data.drop(columns=['–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è_–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å', '–≤—ã—Ä—É—á–∫–∞_—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü'])  
y = merged_data['–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è_–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å']
print(y.unique())
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é, –µ—Å–ª–∏ –æ–Ω–∞ —Å—Ç—Ä–æ–∫–æ–≤–∞—è
if y.dtype == 'object':
    y = y.map({'—Å–Ω–∏–∑–∏–ª–∞—Å—å': 1, '–ø—Ä–µ–∂–Ω–∏–π —É—Ä–æ–≤–µ–Ω—å': 0})
print(y.isna().sum())
print(y.head(10))

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
pipe_final.fit(X_train, y_train)

# –û—Ü–µ–Ω–∏–≤–∞–µ–º –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏
from sklearn.metrics import accuracy_score
y_pred = pipe_final.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(classification_report(y_test, y_pred))
print(y.value_counts(normalize=True))
print("===================")
print(y.dtype)
print(y.unique())
````

````python
# –°–Ω–∞—á–∞–ª–∞ —Å–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–º–µ–Ω–Ω—É—é –¥–ª—è –º–µ—Ç—Ä–∏–∫–∏ recall –¥–ª—è –≤—Å–µ—Ö –º–æ–¥–µ–ª–µ–π
recall_scorer = make_scorer(recall_score, pos_label=1)  # pos_label=1 - –∫–ª–∞—Å—Å —Å–Ω–∏–∂–µ–Ω–∏—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏

# –ö –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π


# –ü–∞–π–ø–ª–∞–π–Ω —Å KNeighborsClassifier
knn_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', KNeighborsClassifier())
])

# –ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
param_grid_knn = {
    'model__n_neighbors': [3, 5, 7, 9],
    'model__weights': ['uniform', 'distance']
}

# –ü–æ–¥–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_search_knn = GridSearchCV(
    knn_pipe,
    param_grid_knn,
    cv=5,
    scoring=recall_scorer,  # –ò—Å–ø–æ–ª—å–∑—É–µ–º recall –≤–º–µ—Å—Ç–æ accuracy
    n_jobs=-1  # –î–ª—è —É—Å–∫–æ—Ä–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
)
grid_search_knn.fit(X_train, y_train)

print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è KNeighborsClassifier:", grid_search_knn.best_params_)
print("–õ—É—á—à–∞—è recall:", grid_search_knn.best_score_)
````

````python
# –ü–∞–π–ø–ª–∞–π–Ω —Å DecisionTreeClassifier
dt_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', DecisionTreeClassifier(random_state=42))
])

# –ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
param_grid_dt = {
    'model__max_depth': [None, 5, 10, 15],
    'model__min_samples_split': [2, 5, 10]
}

# –ü–æ–¥–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_search_dt = GridSearchCV(
    dt_pipe,
    param_grid_dt,
    cv=5,
    scoring=recall_scorer,
    n_jobs=-1
)
grid_search_dt.fit(X_train, y_train)

print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è DecisionTreeClassifier:", grid_search_dt.best_params_)
print("–õ—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å:", grid_search_dt.best_score_)
````

- –ú–æ–∂–Ω–æ —Ä–∞–∑–≥–æ–≤–æ—Ä–∏—Ç—å –º–æ–¥–µ–ª—å–∫—É —Å –ø–æ–º–æ—â—å—é –ø–æ–¥–∞—á–∏ verbose = 3 (–ß—Ç–æ–±—ã –æ—Ç—Å–ª–µ–∂–∏–≤–∞—Ç—å –ø—Ä–æ—Ü–µ—Å—Å—ã –æ–±—É—á–µ–Ω–∏—è –≤ —Ä–µ–∂–∏–º–µ –æ–Ω–ª–∞–π–Ω), –∞ —á—Ç–æ–±—ã –≤—ã–≤–æ–¥–∏–ª–æ—Å—å –Ω–µ —Ç–æ–ª—å–∫–æ –º–µ—Ç—Ä–∏–∫–∞   –ø–æ –∫–æ—Ç–æ—Ä–æ–π —Ç—ã –ø–æ–¥–±–∏—Ä–∞–µ—à—å –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –Ω–æ —Ü–µ–ª—ã–π —Å–ø–∏—Å–æ–∫ –∏–∑ –º–µ—Ç—Ä–∏–∫ (–º–æ–∂–µ—Ç  –¥–æ–±–∞–≤–∏—Ç—å scoring=['–ú–µ—Ç—Ä–∏–∫–∞', '', ''...], –∞ —á–µ—Ä–µ–∑ refit   —É–∫–∞–∑–∞—Ç—å –ø–æ –∫–∞–∫–æ–π –º–µ—Ç—Ä–∏–∫–µ –ø–æ–¥–±–∏—Ä–∞—é—Ç—Å—è –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã )     
     
    

````python
# –ü–∞–π–ø–ª–∞–π–Ω —Å LogisticRegression
lr_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LogisticRegression(max_iter=1000, random_state=RANDOM_STATE))
])

# –ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
param_grid_lr = {
    'model__C': [0.1, 1.0, 10.0],
    'model__solver': ['lbfgs', 'liblinear']
}

# –ü–æ–¥–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_search_lr = GridSearchCV(
    lr_pipe,
    param_grid_lr,
    cv=5,
    scoring=recall_scorer,
    n_jobs=-1
)
grid_search_lr.fit(X_train, y_train)

print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è LogisticRegression:", grid_search_lr.best_params_)
print("–õ—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å:", grid_search_lr.best_score_)
````

````python
# –ü–∞–π–ø–ª–∞–π–Ω —Å SVC
svc_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', SVC(random_state=RANDOM_STATE))
])

# –ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
param_grid_svc = {
    'model__C': [0.1, 1.0, 10.0],
    'model__kernel': ['linear', 'rbf']
}

# –ü–æ–¥–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_search_svc = GridSearchCV(
    svc_pipe,
    param_grid_svc,
    cv=5,
    scoring=recall_scorer,
    n_jobs=-1
)
grid_search_svc.fit(X_train, y_train)

print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è SVC:", grid_search_svc.best_params_)
print("–õ—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å:", grid_search_svc.best_score_)
````

–ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ SHAP, –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –≤–∞–∂–Ω–æ—Å—Ç–∏.

````python
# –ü–æ–ª—É—á–∞–µ–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –∏–∑ GridSearchCV
final_model = grid_search_lr.best_estimator_

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –º–æ–¥–µ–ª—å –ø–∞–π–ø–ª–∞–π–Ω–æ–º
if isinstance(final_model, Pipeline):
    # 1. –ü—Ä–æ–≤–µ—Ä—è–µ–º —à–∞–≥–∏ –ø–∞–π–ø–ª–∞–π–Ω–∞
    print("–®–∞–≥–∏ –ø–∞–π–ø–ª–∞–π–Ω–∞:", final_model.named_steps.keys())
    
    # –ü–æ–ª—É—á–∞–µ–º —à–∞–≥ —Å –º–æ–¥–µ–ª—å—é
    model_step = final_model.named_steps['model']
    print(f"–¢–∏–ø –º–æ–¥–µ–ª–∏: {type(model_step)}")
    
    # –ü–æ–ª—É—á–∞–µ–º –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç—á–∏–∫
    preprocessor = final_model.named_steps['preprocessor']
    
    # 2. –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ
    try:
        X_train_transformed = preprocessor.transform(X_train)
        X_test_transformed = preprocessor.transform(X_test)
        print(f"–†–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: {X_train_transformed.shape}")
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö: {e}")
        # return

    # 3. –ü–æ–ª—É—á–∞–µ–º –∏–º–µ–Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
    feature_names = []
    
    # –î–ª—è ColumnTransformer –∏—Å–ø–æ–ª—å–∑—É–µ–º –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–π –º–µ—Ç–æ–¥
    if hasattr(preprocessor, 'get_feature_names_out'):
        feature_names = preprocessor.get_feature_names_out()
        print("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω get_feature_names_out()")
    else:
        # –†—É—á–Ω–æ–µ –∏–∑–≤–ª–µ—á–µ–Ω–∏–µ –¥–ª—è –∫–∞—Å—Ç–æ–º–Ω—ã—Ö –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π
        if 'ohe' in preprocessor.named_transformers_:
            ohe = preprocessor.named_transformers_['ohe'].named_steps['ohe']
            ohe_features = ohe.get_feature_names_out()
            feature_names.extend(ohe_features)
        
        if 'num_group1' in preprocessor.named_transformers_:
            num_features = preprocessor.named_transformers_['num_group1'].get_feature_names_out()
            feature_names.extend(num_features)
        
        if 'num_group2' in preprocessor.named_transformers_:
            num_features = preprocessor.named_transformers_['num_group2'].get_feature_names_out()
            feature_names.extend(num_features)
    
    print(f"–ü–æ–ª—É—á–µ–Ω–æ {len(feature_names)} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
    print("–ü—Ä–∏–º–µ—Ä—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:", feature_names[:5])
    
    # –°–æ–∑–¥–∞–µ–º DataFrame
    X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=feature_names)
    X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=feature_names)
    
    # 4. –í—ã–±–∏—Ä–∞–µ–º –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –æ–±—ä—è—Å–Ω–∏—Ç–µ–ª—å
    try:
        # –î–ª—è –ª–∏–Ω–µ–π–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π
        if isinstance(model_step, (LogisticRegression, LinearRegression)):
            explainer = shap.LinearExplainer(model_step, X_train_transformed_df)
            print("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω LinearExplainer")
        
        # –î–ª—è tree-based –º–æ–¥–µ–ª–µ–π
        elif isinstance(model_step, (DecisionTreeClassifier, RandomForestClassifier, GradientBoostingClassifier)):
            explainer = shap.TreeExplainer(model_step)
            print("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω TreeExplainer")
        
        # –î–ª—è –¥—Ä—É–≥–∏—Ö –º–æ–¥–µ–ª–µ–π
        else:
            explainer = shap.Explainer(model_step.predict, X_train_transformed_df)
            print("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω –æ–±—â–∏–π Explainer")
        
        # –í—ã—á–∏—Å–ª—è–µ–º SHAP-–∑–Ω–∞—á–µ–Ω–∏—è
        shap_values = explainer(X_test_transformed_df)
        
        # 5. –ü—Ä–æ–≤–µ—Ä—è–µ–º SHAP-–∑–Ω–∞—á–µ–Ω–∏—è
        print(f"SHAP values shape: {shap_values.shape}")
        print(f"Min SHAP value: {np.min(shap_values.values)}")
        print(f"Max SHAP value: {np.max(shap_values.values)}")
        
        # 6. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è

        plt.figure(figsize=(30, 4))

        shap.summary_plot(shap_values, X_test_transformed_df, plot_type="bar", max_display=15, show=False)

        plt.title("–ì–ª–æ–±–∞–ª—å–Ω–∞—è –≤–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")

        plt.tight_layout()

        plt.show()
        
        plt.figure(figsize=(30, 8))

        plt.rcParams['ytick.labelsize'] = 5

        shap.summary_plot(shap_values, X_test_transformed_df, max_display=15, show=False)

        plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ SHAP-–∑–Ω–∞—á–µ–Ω–∏–π")

        plt.tight_layout()

        plt.show()
        
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ SHAP: {e}")
        import traceback
        traceback.print_exc()

else:
    print("–ú–æ–¥–µ–ª—å –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ø–∞–π–ø–ª–∞–π–Ω–æ–º, —Ç—Ä–µ–±—É–µ—Ç—Å—è –¥—Ä—É–≥–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞")

# –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É waterfall
shap.plots.waterfall(shap_values[0])
````
