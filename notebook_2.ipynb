````python
# Функция для очистки данных
def clean_data(df):
    # Удаляем лишние пробелы в текстовых столбцах
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].str.strip()
    
    # Преобразуем текстовые значения к нижнему регистру
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].str.lower()
    
    return df

# Очищаем каждый DataFrame в списке
for i, data in enumerate(tables_list):
    tables_list[i] = clean_data(data)

# Поиск явных дубликатов
print("Поиск явных дубликатов:")
for data in tables_list:
    print(f"Количество явных дубликатов: {data.duplicated().sum()}")

# Поиск скрытых дубликатов по категориальным столбцам
print("\nПоиск скрытых дубликатов по категориальным столбцам:")
for data in tables_list:
    categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()
    
    # Проверяем, есть ли категориальные столбцы
    if len(categorical_columns) > 0:
        print(f"Количество скрытых дубликатов: {data.duplicated(subset=categorical_columns).sum()}")
    else:
        print("Нет категориальных столбцов для проверки дубликатов.")

# Поиск пропусков
print("\nПоиск пропусков:")
for data in tables_list:
    print(data.isna().sum())

# Обзор уникальных значений
print("\nОбзор уникальных значений:")
for data in tables_list:
    print(data.nunique())
````

````python


market_file_num_columns = market_file.drop(columns='id', errors='ignore') \
                                   .select_dtypes(include=['int64', 'float64']).columns.tolist()
market_money_num_columns = market_money.drop(columns='id', errors='ignore') \
                                   .select_dtypes(include=['int64', 'float64']).columns.tolist()
market_time_num_columns = market_time.drop(columns='id', errors='ignore') \
                                   .select_dtypes(include=['int64', 'float64']).columns.tolist()
market_file_cat_columns = market_file.drop(columns='id', errors='ignore') \
                                   .select_dtypes('object').columns.tolist()
market_money_cat_columns = market_money.drop(columns='id', errors='ignore') \
                                   .select_dtypes('object').columns.tolist()
market_time_cat_columns = market_time.drop(columns='id', errors='ignore') \
                                   .select_dtypes('object').columns.tolist()
# Функция для построения гистограммы и диаграммы размаха для столбцов таблиц
def histogram_boxplot(table_file, label):
    label_common = label.replace('_', ' ').capitalize()
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
    sns.histplot(data=table_file, x=label, ax=axs[0], bins = 40, kde=True)
    axs[0].set_xlabel(label_common)
    axs[0].set_ylabel('Количество')
    axs[0].set_title('Гистограмма')
    sns.boxplot(data=table_file, x=label, ax=axs[1])
    axs[1].set_xlabel(label_common)
    axs[1].set_title('Диаграмма размаха')
    fig.suptitle(label_common)
    plt.show()

def pie_plot(table_file, label):
    # label_common = label.replace('_', ' ').capitalize()
    # fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(12, 4))
    # table_file[label].value_counts().plot(y = 'count', kind = 'pie', autopct='%1.0f%%', title = label_common, ax = axs)
    # plt.show()
    palette_color = sns.color_palette('bright') 
    table_file[label].value_counts().plot(y='count', kind='pie', autopct='%1.0f%%', title=label.capitalize(), colors=palette_color, figsize=(14, 6))
    plt.show()


print("Список количественных столбцов без 'id' в market_file:")
print(market_file_num_columns)
print("Список количественных столбцов без 'id' в market_money:")
print(market_money_num_columns)
print("Список количественных столбцов без 'id' в market_time:")
print(market_time_num_columns)
print("Список категориальных столбцов без 'id' в market_file:")
print(market_file_cat_columns)
print("Список категориальных столбцов без 'id' в market_money:")
print(market_money_cat_columns)
print("Список категориальных столбцов без 'id' в market_time:")
print(market_time_cat_columns)
````

#### 6.2 Добавляем ohe кодирование категориальных данных - создаём пайплайн

````python
ohe_pipe = Pipeline(
    [
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore')
        )
    ]
) 
ord_pipe = Pipeline(
    [
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['нет', 'да'],
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
    ]
)
````

Объединяем все признаки в ColumnTransformer

````python

# Разделяем категориальные признаки на 2 группы - для разных скейлеров
ohe_features = ['тип_сервиса', 'популярная_категория']  # Все категории для OneHotEncoder
ord_features = ['разрешить_сообщать']  # Одна категория для OrdinalEncoder

# Создаем ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', ohe_pipe, ohe_features),  # Используем пайплайн для OneHotEncoder
        ('ord', ord_pipe, ord_features)   # Используем пайплайн для OrdinalEncoder
    ],
    remainder='drop'
)
# Разделяем кколичественные признаки на 2 группы - для разных скейлеров
num_features_group1 = ['выручка_препредыдущий_месяц', 'выручка_предыдущий_месяц', 'длительность']
num_features_group2 = ['акционные_покупки', 'маркет_актив_6_мес', 'маркет_актив_тек_мес']

# Создаем ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', ohe_pipe, ohe_features),
        ('num_group2', StandardScaler(), num_features_group2),
        ('num_group1', MinMaxScaler(), num_features_group1)
    ],
    remainder='drop'
)
````

## 7. Анализ важности признаков
Используем метод SHAP, построим график важности.

````python
# Получаем лучшую модель из GridSearchCV
final_model = grid_search_lr.best_estimator_

# Проверяем, является ли модель пайплайном
if isinstance(final_model, Pipeline):
    # 1. Проверяем шаги пайплайна
    print("Шаги пайплайна:", final_model.named_steps.keys())
    
    # Получаем шаг с моделью
    model_step = final_model.named_steps['model']
    print(f"Тип модели: {type(model_step)}")
    
    # Получаем предобработчик
    preprocessor = final_model.named_steps['preprocessor']
    
    # 2. Преобразуем данные
    try:
        X_train_transformed = preprocessor.transform(X_train)
        X_test_transformed = preprocessor.transform(X_test)
        print(f"Размерность преобразованных данных: {X_train_transformed.shape}")
    except Exception as e:
        print(f"Ошибка преобразования данных: {e}")
        # return

    # 3. Получаем имена признаков после преобразования
    feature_names = []
    
    # Для ColumnTransformer используем встроенный метод
    if hasattr(preprocessor, 'get_feature_names_out'):
        feature_names = preprocessor.get_feature_names_out()
        print("Использован get_feature_names_out()")
    else:
        # Ручное извлечение для кастомных преобразований
        if 'ohe' in preprocessor.named_transformers_:
            ohe = preprocessor.named_transformers_['ohe'].named_steps['ohe']
            ohe_features = ohe.get_feature_names_out()
            feature_names.extend(ohe_features)
        
        if 'num_group1' in preprocessor.named_transformers_:
            num_features = preprocessor.named_transformers_['num_group1'].get_feature_names_out()
            feature_names.extend(num_features)
        
        if 'num_group2' in preprocessor.named_transformers_:
            num_features = preprocessor.named_transformers_['num_group2'].get_feature_names_out()
            feature_names.extend(num_features)
    
    print(f"Получено {len(feature_names)} признаков")
    print("Примеры признаков:", feature_names[:5])
    
    # Создаем DataFrame
    X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=feature_names)
    X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=feature_names)
    
    # 4. Выбираем правильный объяснитель
    try:
        # Для линейных моделей
        if isinstance(model_step, (LogisticRegression, LinearRegression)):
            explainer = shap.LinearExplainer(model_step, X_train_transformed_df)
            print("Использован LinearExplainer")
        
        # Для tree-based моделей
        elif isinstance(model_step, (DecisionTreeClassifier, RandomForestClassifier, GradientBoostingClassifier)):
            explainer = shap.TreeExplainer(model_step)
            print("Использован TreeExplainer")
        
        # Для других моделей
        else:
            explainer = shap.Explainer(model_step.predict, X_train_transformed_df)
            print("Использован общий Explainer")
        
        # Вычисляем SHAP-значения
        shap_values = explainer(X_test_transformed_df)
        
        # 5. Проверяем SHAP-значения
        print(f"SHAP values shape: {shap_values.shape}")
        print(f"Min SHAP value: {np.min(shap_values.values)}")
        print(f"Max SHAP value: {np.max(shap_values.values)}")
        
        # 6. Визуализация

        plt.figure(figsize=(20, 4))

        shap.summary_plot(shap_values, X_test_transformed_df, plot_type="bar", max_display=15, show=False)

        plt.title("Глобальная важность признаков")

        plt.tight_layout()

        plt.show()
        
        plt.figure(figsize=(20, 8))

        shap.summary_plot(shap_values, X_test_transformed_df, max_display=15, show=False)

        plt.title("Распределение SHAP-значений")

        plt.tight_layout()

        plt.show()
        
    except Exception as e:
        print(f"Ошибка при вычислении SHAP: {e}")
        import traceback
        traceback.print_exc()

else:
    print("Модель не является пайплайном, требуется другая обработка")

# Построим диаграмму waterfall
shap.plots.waterfall(shap_values[0])
````
